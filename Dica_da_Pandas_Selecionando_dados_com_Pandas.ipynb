# Qual a diferena entre usar loc e usar apenas colchetes para filtrar colunas em Pandas?

```python
import pandas as pd
```

```python
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
```

```python
iris
```

```python
iris['petal_length']
```

```python
iris[iris['petal_length']>5]
```

```python
iris.loc[iris['petal_length']>5]
```

```python
iris[iris['species']=='setosa']['petal_length']
```

```python
iris.loc[iris['species']=='setosa','petal_length']
```

```python
iris[iris['species']=='setosa']['petal_length'] = 1
```

```python
iris.loc[iris['species']=='setosa','petal_length'] = 1
```

```python
iris
```

```python
iris[iris['species']=='setosa']
```

```python
iris[iris['species']=='virginica']
iris.loc[iris['species']=='virginica']

iris[iris['species']=='virginica']['sepal_length']
iris.loc[iris['species']=='virginica','sepal_length']

iris[iris['sepal_length']>7]

iris.filter(items=['sepal_length', 'sepal_width'])
iris[['sepal_length', 'sepal_width']]
```
